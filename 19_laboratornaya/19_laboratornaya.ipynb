## 1. Fight

```python
import random


class Warrior:

    def __init__(self, name):
        self.name = name
        self.health = 100

        
player_1 = Warrior('первый воин')
player_2 = Warrior('второй воин')

while player_1.health > 0 and player_2.health > 0:
    x = random.randint(0, 1)
    if x == 1:
        player_1.health -= 20
        print('Атаковал второй воин. У первого воина осталось здоровья:', player_1.health)
    else:
        player_2.health -= 20
        print('Атаковал первый воин. У второго юнита осталось здоровья:', player_2.health)

if player_1.health > 0:
    print('\nПобедил', player_1.name)
else:
    print('\nПобедил', player_2.name)
```

## 3. Circle

```python
class Circle:
    pi = 3.1415926
    
    def __init__(self, x_coord, y_coord, radius):
        self.x_coord = x_coord
        self.y_coord = y_coord
        self.radius = radius

    def area(self):
        return self.pi * self.radius ** 2

    def perimeter(self):
        return 2 * self.pi * self.radius

    def increase(self, coefficient):
        self.radius *= coefficient

    def intersection(self, other):
        if (self.x_coord - other.x_coord) ** 2 + (self.y_coord - other.y_coord) ** 2 <= (
                self.radius + other.radius) ** 2:
            print('Окружности пересекаются')
        else:
            print('Окружности не пересекаются')


circle_1 = Circle(0, 0, 1)
circle_2 = Circle(1, 1, 2)

print('Площадь круга:', circle_1.area())
print('Периметр окружности:', circle_1.perimeter())
circle_1.increase(2)
print('Новый радиус окружности:', circle_1.radius)
circle_1.intersection(circle_2)
```

## 4. Fathers, mothers and children

```python
class Parent:

    def __init__(self, name, age, children_list):
        self.name = name
        self.age = age
        self.children_list = children_list

    def information(self):
        return self.name, self.age, self.children_list
    
    def calm_the_child_down(self, child):
        if child.name in self.children_list:
            if child.state_of_calm == 'нервничает':
                print('Вы поддержали ребенка.', child.name, 'снова спокойна.' )
            elif child.state_of_calm == 'плачет':
                print('Вы успокоили ребенка.', child.name, 'снова спокойна.')
            elif child.state_of_calm == 'боится':
                print('Вы успокоили ребенка.', child.name, 'снова спокойна.')
            else:
                print('Ребенок спокойный. Успокаивать не нужно.')
        else:
            print('Это не мой ребенок.')

    def feed_the_child(self, child):
        if child.name in self.children_list:
            if child.state_of_hunger == ('голодный' or 'голодная'):
                print(child.name, 'вкусно накормлен(а).')
        else:
            print('Это не мой ребенок.')


class Child:

    def __init__(self, name, age, state_of_calm, state_of_hunger):
        self.name = name
        self.age = age
        self.state_of_calm = state_of_calm
        self.state_of_hunger = state_of_hunger

    def check(self, parent):
        if self.age + 16 > parent.age:
            print('Возраст должен быть меньше возраста родителя хотя бы на 16 лет.')
            return 0
        else:
            return 1
            
            
parent_1 = Parent('Мария', 40, ['Анна', 'Ксения', 'Вера'])
child_1 = Child('Анна', 21, 'нервничает', 'не голодная')
child_2 = Child('Ксения', 16, 'нервничает', 'не голодная')
child_3 = Child('Вера', 0, 'плачет', 'голодная')

print('Информация о родителе:', parent_1.information())
if child_1.check(parent_1) == 1:
    print(child_1.name, child_1.state_of_calm)
    parent_1.calm_the_child_down(child_1)
    print(child_1.name, child_1.state_of_hunger)
    parent_1.feed_the_child(child_1)

if child_1.check(parent_1) == 1:
    print(child_2.name, child_2.state_of_calm)
    parent_1.calm_the_child_down(child_2)
```

## 5. Fun Farm

```python
class Gardener:

    def __init__(self, name, garden_bed):
        self.name = name
        self.garden_bed = garden_bed

    def take_care(self):
        return self.garden_bed.grow()

    def harvest(self):
        return self.garden_bed.harvest()


class Potato:

    def __init__(self, number, stage_of_maturity):
        self.number = number
        self.stage_of_maturity = stage_of_maturity

    def information(self):
        return self.number, self.stage_of_maturity

    def grow(self):
        if self.stage_of_maturity + 1 <= 4:
            self.stage_of_maturity += 1
        else:
            self.stage_of_maturity = 1


class Garden_bed:

    def __init__(self, potato_list):
        self.potato_list = potato_list

    def information(self):
        return [potato.stage_of_maturity for potato in self.potato_list]

    def grow(self):
        return [potato.grow() for potato in self.potato_list]

        
        
potato_1 = Potato(1, 1)
potato_2 = Potato(2, 4)
potato_3 = Potato(3, 3)
potato_4 = Potato(4, 2)
potato_5 = Potato(5, 1)
garden_bed_1 = Garden_bed([potato_1, potato_2, potato_3, potato_4, potato_5])
gardener_1 = Gardener('Gardener1', garden_bed_1)

print(garden_bed_1.information())
gardener_1.take_care()
print(garden_bed_1.information())
gardener_1.take_care()
print(garden_bed_1.information())
gardener_1.take_care()
print(garden_bed_1.information())
```
