## 1. Taxes

```python
class Property:

    def __init__(self, worth):
        self.worth = worth


class Apartment(Property):

    def __init__(self, worth):
        super().__init__(worth)

    def calculation(self):
        return self.worth / 1000


class Car(Property):

    def __init__(self, worth):
        super().__init__(worth)

    def calculation(self):
        return self.worth / 200


class CountryHouse(Property):

    def __init__(self, worth):
        super().__init__(worth)

    def calculation(self):
        return self.worth / 500
    
    
salary = float(input('Введите количество Ваших денег: '))
apartment = Apartment(float(input('Введите стоимость квартиры: ')))
car = Car(float(input('Введите стоимость машины: ')))
house = CountryHouse(float(input('Введите стоимость дома: ')))
print('Суммарный налог:', round(apartment.calculation() + car.calculation() + house.calculation(), 2))
if apartment.calculation() + car.calculation() + house.calculation() > salary:
    print('Не хватает на погашение налогов:', round(apartment.calculation() + car.calculation()
                                                    + house.calculation() - salary, 2))
```

## 2. Karma

```python
import random


class Error(Exception):
    pass


class KillError(Error):
    pass


class DrunkError(Error):
    pass


class CarCrashError(Error):
    pass


class GluttonyError(Error):
    pass


class DepressionError(Error):
    pass


class Life:
    max_karma = 500

    def __init__(self):
        self.karma = 0

    def one_day(self):
        with open('karma.log', 'a') as ouf:
            try:
                self.karma += random.randint(1, 7)
                if 10 == random.randint(1, 10):
                    raise random.choice([KillError, DrunkError, CarCrashError, GluttonyError, DepressionError])
            except KillError:
                ouf.write('KillError\n')
            except DrunkError:
                ouf.write('DrunkError\n')
            except CarCrashError:
                ouf.write('CarCrushError\n')
            except GluttonyError:
                ouf.write('GluttonyError\n')
            except DepressionError:
                ouf.write('DepressionError\n')

                
life = Life()
while life.karma < life.max_karma:
    life.one_day()
    print(life.karma)
```

## 3. Your dictionary

```python
class MyDict:
    
    def __init__(self, dictionary):
        self.dict = dictionary
        
    def check(self, key):
        if self.dict.get(key) == None: 
            return 0
        else: 
            return self.dict.get(key)
        
        
test_dict = {1: 'a', 2 : 'b', 3 : 'c', 4 : 'd' }

my_dict = MyDict(test_dict)
print(my_dict.check(4))
print(my_dict.check(5))
```

## 4. Company

```python
class Person:
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = int(age)


class Employee(Person):

    def __init__(self, name, surname, age):
        super().__init__(name, surname, age)

    def salary(self):
        pass


class Manager(Employee):

    def __init__(self, name, surname, age):
        super().__init__(name, surname, age)

    def salary(self):
        return 13000


class Agent(Employee):

    def __init__(self, name, surname, age, sales_volume):
        super().__init__(name, surname, age)
        self.sales_volume = int(sales_volume)

    def salary(self):
        return 5000 + self.sales_volume * 0.05


class Worker(Employee):

    def __init__(self, name, surname, age, time):
        super().__init__(name, surname, age)
        self.time = int(time)

    def salary(self):
        return 100 * self.time


list_of_employees = [Manager('Пётр', 'Степанов', 47), 
                     Manager('Владимир', 'Пономарев', 41), 
                     Manager('Алексей', 'Шутов', 45), 
                     Agent('Андрей', 'Пушилин', 37, 10000), 
                     Agent('Тимур', 'Рясов', 32, 8000), 
                     Agent('Дмитрий', 'Касимов', 34, 7500), 
                     Worker('Ольга', 'Святкина', 29, 36), 
                     Worker('Константин', 'Гинс', 27, 36), 
                     Worker('Ирина', 'Лаврова', 30, 40)]

sum_salary = 0
for elem in list_of_employees:
    sum_salary += elem.salary()
print('Заработная плата всех служащих:', sum_salary)
```

## 5. The car!

```python
import math


class Car:
    
    def __init__(self, x, y, angle):
        self.x = x
        self.y = y
        self.angle = angle

    def move(self, distance, angle):
        self.x += distance * math.cos(angle)
        self.y += distance * math.sin(angle)

    def info(self):
        print(self.x, self.y)


class Bus(Car):
    def __init__(self, x, y, angle, passengers, ticket_cost):
        super().__init__(x, y, angle)
        assert passengers > 0, 'Число пассажиров не может быть меньше 0'
        self.passengers = passengers
        self.ticket_cost = ticket_cost
        self.money = self.passengers * self.ticket_cost

    def come_in(self, people):
        self.passengers += people
        self.money += people * self.ticket_cost

    def come_out(self, people):
        self.passengers -= people
        assert self.passengers > 0, 'Число пассажиров не может быть меньше 0'

    def move(self, distance, angle):
        self.x += distance * math.cos(angle)
        self.y += distance * math.sin(angle)

    def get_info(self):
        print(f'Число пассажиров: {self.passengers}\nКоличество полученных денег: {self.money}')
        
        
bus = Bus(0, 0, 0, 50, 2)
bus.move(20, 3.14)
bus.info()
bus.get_info()
bus.come_in(5)
bus.come_out(3)
bus.move(40, 3.14/60)
bus.info()
bus.get_info()
```

## 7. Stack

```python
class LIFO:

    def __init__(self):
        self.storage = []

    def push(self, item):
        self.storage.append(item)

    def pop(self):
        self.storage.pop(-1)

    def info(self):
        print(self.storage)


class TaskManager():

    def __init__(self):
        self.storage = {}

    def new_task(self, item: str, priority: int):
        if priority not in self.storage:
            self.storage[priority] = [item]
        else:
            self.storage[priority].append(item)
        lst_keys = list(self.storage.keys())
        lst_keys.sort()
        dct = {}
        for key in lst_keys:
            if key not in dct:
                dct[key] = self.storage[key]
            else:
                dct[key].append(self.storage[key])
        self.storage = dct

    def delete_task(self, item: str, priority: int):
        if len(self.storage[priority]) == 1:
            del self.storage[priority]
        else:
            for i in range(len(self.storage[priority])):
                if self.storage[priority][i] == item:
                    index = i
            del self.storage[priority][index]

    def info(self):
        for key in self.storage:
            print(key, end='. ')
            if len(self.storage[key]) == 1:
                print(self.storage[key][0])
            else:
                for i in range(len(self.storage[key]) - 1):
                    print(self.storage[key][i], end='; ')
                print(self.storage[key][len(self.storage[key]) - 1])


lifo = LIFO()
lifo.push(1)
lifo.push(2)
lifo.info()
lifo.pop()
lifo.info()

print()

manager = TaskManager()
manager.new_task('сделать уборку', 4)
manager.new_task('помыть посуду', 4)
manager.new_task('отдохнуть', 1)
manager.new_task('поесть', 2)
manager.new_task('сдать дз', 2)
manager.info()
print()
manager.delete_task('сдать дз', 2)
manager.info()
```
