## 1. Squares of numbers

```python
class SimpleIterator:
    
    def __iter__(self):
        return self
    
    def __init__(self, limit):
        self.limit = limit
        self.counter = 0
        
    def __next__(self):
        if self.counter < self.limit:
            self.counter += 1
            return self.counter ** 2
        else:
            raise StopIteration
            
def simple_generator(N):
    for elem in range(1, N+1):
        yield elem ** 2

def generate_expression(N):
    return [elem**2 for elem in range(1, N+1)]
        
        
N = int(input('N = '))

iter_1 = SimpleIterator(N)
print('Класс-итератор:')
for i in iter_1:
    print(i, end=' ')

print('\n\nФункция-генератор:')
for i in simple_generator(N):
    print(i, end=' ')

print('\n\nГенераторное выражение:')
for i in generate_expression(N):
    print(i, end=' ')
```

## 2. Refactoring

```python
def generator(lst_1, lst_2):
    for x in lst_1:
        for y in lst_2:
            yield x, y

            
list_1 = [2, 5, 7, 10]
list_2 = [3, 8, 4, 9]
to_find = 56
for x, y in generator(list_1, list_2):
    result = x * y
    print(x, '*', y, '=', result)
    if x * y == to_find: 
        print('Found!!!')
        break
```

## 4. The Hofstadter sequence

```python
def QHofstadter(start):
    if start != [1, 1]:
        return 1
    seq = start[:]
    while 1:
        q = seq[-seq[-1]] + seq[-seq[-2]]
        seq.append(q)
        yield q

i = 1
for x in QHofstadter([1, 1]):
    if i <= 20:
        print(x, end=' ')
        i += 1
    else:
        break
```

## 5. Number of rows

```python
import os


def lines(directory, path):
    num = 0
    for elem in os.listdir(directory):
        if os.path.isfile(path + elem) and elem != 'karma.log':
            for line in open(path + elem, 'r', encoding='utf-8'):
                if line != '\n' and not line.startswith('#'):
                    num += 1
            yield num


directory = 'C:/Users/АВА/Desktop/Python/20_laboratornaya'
path = directory + '/'
summa = 0
for number_of_lines in lines(directory, path):
    summa += number_of_lines
print('Общее число строк кода в 20 лабораторной:', summa)
```
